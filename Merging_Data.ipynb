## Import librairies :

```python
import os
import pandas as pd
import zipfile
from google.colab import files
```

## Merge scraped dataset

```python
#  the ZIP file path and destination folder :
file_zip = '/content/data of scrapping texts.zip'
destination_folder = '/content/data of scrapping texts_UnZiped'

# Unzip the ZIP file :
with zipfile.ZipFile(file_zip, 'r') as zip_ref:
    zip_ref.extractall(destination_folder)


```

```python
#the path to the folder containing the unzipped CSV files
folder_csv = '/content/data of scrapping texts_UnZiped/data of scrapping texts/data'

# Create a list to store the DataFrames of each CSV file
dfs = []

# Browse each file in the folder
for fichier_csv in os.listdir(folder_csv):
    if fichier_csv.endswith('.csv'):
        path = os.path.join(folder_csv, fichier_csv)
        df = pd.read_csv(path)
        dfs.append(df)

# Concatenate all DataFrames into one
dataset_all = pd.concat(dfs, ignore_index=True)

# Save the result in a new CSV file
dataset_all.to_csv('/content/Dataset.csv', index=False)

```
